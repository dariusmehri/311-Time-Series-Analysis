```python
import pandas as pd
import numpy as np

path = "C:\\Users\\dmehri\\Documents\\DATA\\311 Analysis\\"

df0 = pd.read_csv(path + "311_Service_Requests_2012_small.csv", low_memory=False)
df0["Year"] = "2012"
df1 = pd.read_csv(path + "311_Service_Requests_2013_small.csv", low_memory=False)
df1["Year"] = "2013"
df2 = pd.read_csv(path + "311_Service_Requests_2014_small.csv", low_memory=False)
df2["Year"] = "2014"
df3 = pd.read_csv(path + "311_Service_Requests_2015_small.csv", low_memory=False)
df3["Year"] = "2015"
df4 = pd.read_csv(path + "311_Service_Requests_2016_small.csv", low_memory=False)
df4["Year"] = "2016"
df5 = pd.read_csv(path + "311_Service_Requests_2017_small.csv", low_memory=False)
df5["Year"] = "2017"
df6 = pd.read_csv(path + "311_Service_Requests_2018_small.csv", low_memory=False)
df6["Year"] = "2018"
df7 = pd.read_csv(path + "311_Service_Requests_2019_small.csv", low_memory=False)
df7["Year"] = "2019"
df8 = pd.read_csv(path + "311_Service_Requests_2020_small.csv", low_memory=False)
df8["Year"] = "2020"
df9 = pd.read_csv(path + "311_Service_Requests_2021_small.csv", low_memory=False)
df9["Year"] = "2021"

frames = [df0, df1, df2, df3, df4, df5, df6, df7, df8, df9]

df = pd.concat(frames)
df = df.reset_index(drop=True)
```

```python
df.to_csv(path + "311_Service_Requests_2012-2021", index=False)
```

```python
df.dtypes
```

```python
df2 = df[["Year"]]
df2["Number of Complaints"] = 1

df2 = df2.groupby(['Year']).sum()

df2 = df2.add_suffix('').reset_index()
```

```python
df2
```

```python

df2.to_csv(path + "311 Complaints by Year.csv", index=False)
```
